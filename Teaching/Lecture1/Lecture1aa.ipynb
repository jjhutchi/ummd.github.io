## The COVID Dataset

Data form 200 markets

* Sales: thousands of units sold
* TV: amount spent on TV advertising (in thousands of dollars)
* radio: amount spent on radio advertising (in thousands of dollars)
* newspaper: amount spent on newspaer advertising (in thousands of dollars)

### Load a Dataset

```r
COVID <- read.csv("https://data.ontario.ca/dataset/f4112442-bdc8-45d2-be3c-12efae72fb27/resource/455fd63b-603d-4608-8216-7d8647f43350/download/conposcovidloc.csv")
COVID
```

### Attach Variable Names to Workspace

```r
attach(COVID)


COVID$Accurate_Episode_Date <- as.Date(COVID$Accurate_Episode_Date, format="%Y-%m-%d") #tell R that this is a Date format, not just a string of characters, this will allow us to use some special functions on it
```

### Plot Data

```r
totalCases <- table(cut(COVID$Accurate_Episode_Date, 'day')) #try changing 'day' to 'month'
totalCases
```

```r
plot(totalCases)
plot(as.data.frame(totalCases))
```

### Fitting a Model to this data

Store this Table in a new Data Frame

```r
datanumeric <- as.data.frame(totalCases)
```

Now convert the date into a numeric value

```r
datanumeric$Var1 <- as.numeric(datanumeric$Var1) 
```

And fit a linear model to the data

```r
lmfit <- lm(datanumeric) # this fits the model  
summary(lmfit) #this prints a summary of the model fit

plot(as.data.frame(totalCases)) # plot the data again
abline(lmfit, col="red") # plot the fitted line from the model fit
```

What about men vs. women?

```r
totalCasesMale <- table(cut(COVID$Accurate_Episode_Date[COVID$Client_Gender == "MALE"], 'day'))
totalCasesFemale <- table(cut(COVID$Accurate_Episode_Date[COVID$Client_Gender == "FEMALE"], 'day'))

sum(totalCasesMale)
```
