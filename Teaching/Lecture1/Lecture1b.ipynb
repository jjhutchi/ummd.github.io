## Now use ggplot

```r
require(ggplot2)
df <- data.frame(Education,Income) #put the data into a dataframe so that ggplot can use it properly
```

ggplot uses "layers" which you add with the "+" command.

```r
g <- ggplot(df, aes(x=Education, y=Income)) + geom_point()  #create an axis with aes(), then add a scatterplot with geom_point()
g   # print the plot to screen
```

Now plot a layer of the true DGP

```r
y1 <- 60*pnorm((Education-16)/2.5)+20 #this is the true DGP
g <- g + geom_line(aes(y=y1), colour="blue") # add it to the plot using geom_line()
g
```
