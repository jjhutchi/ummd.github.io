# Lecture 2D - The Bayes Error Rate

These examples will give intuition for the Bayes Error Rate

The true DGP generates binary data with the probability given by the black line below

```r
x <- seq(-4,4, by=0.1)

b1 <- 2

p <- exp(b1*x)/(1+exp(b1*x))

plot(x,p)
lines(x,p,col='black')


n <- 810
y <- rbinom(n,1,rep(p,10))
points(rep(x,10),y,col=y+2)
abline(h=0.5)
```

```r
yh <- p>0.5 #the bayes optimal predictor is to always predict the class with the largest probability. For a binary variable, this cutoff is 0.5.
plot(x,yh)
```

How many errors will we make, assuming we knew the true model?

```r
error_rate <- mean(y!=yh)
error_rate
```

Now let's repeat, but with a true model that has more variability. This means the true probaility is "flatter" than before. So at each x, the true probability will be further from our prediction (either 0 or 1).

```r
b1 <- .5

p <- exp(b1*x)/(1+exp(b1*x))

plot(x,p,ylim=c(0,1))
lines(x,p,col='black')

n <- 810
y <- rbinom(n,1,rep(p,10))
points(rep(x,10),y,col=y+2)
abline(h=0.5)
```

```r

error_rate <- mean(y!=yh)
error_rate
```

Because there is more variability, we will make more errors, even if we knew the true model.

## Why is the Bayes Optimal Predictor Optimal?

The Bayes Optimal prediction is to predict the class with the largest estimated probability.

What if we used a different prediction method? Suppose instead of the Bayes Optimal predictor, we randomized our prediction according to our predicted probability? So if the probability were 0.6, we predicted a "1" 60% of the time? Here, we will even assume that we know the true probability!

```r
yh2 <- rbinom(n,1,p)
error_rate <- mean(y!=yh2)
error_rate
```

Whoa, that is way worse. Turns out that this guy Bayes knew something!
